```python
# f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

# 1. Определить корни
# 2. Найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает
# 4. Построить график
# 5. Вычислить вершину
# 6. Определить промежутки, на котором f > 0
# 7. Определить промежутки, на котором f < 0
```

```python
from sympy import *
```

```python
x = Symbol('x')
#func = -12*pow(x, 4)*sin(cos(x)) - 18*pow(x, 3)+5*pow(x, 2) + 10*x - 30
func = pow(x, 3)-3*pow(x, 2)+4
simplify(func)
```

```python
eq_roots = list(nonlinsolve([func], [x]))
print(f'1. Корни уравнения {func}:\nx1 = {eq_roots[0][0]}, x2 = {eq_roots[1][0]}')
```

```python
func_diff = diff(func) # производная уравнения
simplify(func_diff)
```

```python
func_up = solve(func_diff > 0)
func_down = solve(func_diff < 0)
print(f'2. функция {func} возрастает в интервалах {func_up}')
print(f'3. функция {func} убывает в интервалах {func_down}')
```

```python
print(f'4. График функции {func}:')
graph_func = plot(func, (x, -2, 3), line_color='red', title=f'{func} + производная ', show=False)
graph_diff = plot(func_diff, (x, -2, 3), line_color='blue', show=False)
graph_func.append(graph_diff[0])
graph_func.show()
```

```python
eq_roots_minmax = list(nonlinsolve([func_diff], [x]))
print(f'5. Точка максимума функции {func}: [{eq_roots_minmax[0][0]}; {func.subs(x, eq_roots_minmax[0][0])}]')
print(f'   Точка минимума функции {func}: [{eq_roots_minmax[1][0]}; {func.subs(x, eq_roots_minmax[1][0])}]')
```

```python
from random import randint as rand
points = []
points.append(rand(eq_roots[0][0] - 1, eq_roots[0][0]))
points.append(rand(eq_roots[0][0] + 1, eq_roots[1][0]))
points.append(rand(eq_roots[1][0] + 1, eq_roots[1][0] + 2))
intervals = [f'-{oo};{eq_roots[0][0]}', f'{eq_roots[0][0]}; {eq_roots[1][0]}', f'{eq_roots[0][0]}; {oo}']
intervals_values = []
intervals_values.append(func.subs(x, points[0]))
intervals_values.append(func.subs(x, points[1]))
intervals_values.append(func.subs(x, points[2]))
print(intervals_values)
print(intervals)
```

```python
print(f'6, 7: промежутки, на которых функция {func} отрицательна и положительна:')
for i in range(len(intervals)) :
    if intervals_values[i] < 0: print(f'На промежутке ({intervals[i]}) функция < 0')
    else: print(f'На промежутке ({intervals[i]}) функция > 0')
```
